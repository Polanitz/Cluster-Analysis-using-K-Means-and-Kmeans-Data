```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("kmeans.csv")
```

```python
df.head()
```

# K = 2

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df)
```

```python
kmeans.predict(np.array([1,2,3,4,1,2,3]).reshape(1,-1))
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_[:10]
```

```python
plt.scatter(df['C'],df['B'],c=kmeans.labels_,cmap='rainbow')
```

```python
df['labels'] = kmeans.labels_
```

```python
df.head(3)
```

```python
sns.pairplot(df, hue='labels',vars=['A','B','C','D','E','F','G'])
```
